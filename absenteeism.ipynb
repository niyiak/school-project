```python
num_feat = list(df.drop(columns=["ID","Reason for absence","Reason about Diseases","Reason about Pregnancy & Childbirth","Reason about Poisoning & Illness","Light Reasons","Month of absence","Day of the week","Seasons","Hit target","Disciplinary failure","Education","Social drinker","Social smoker","Weight","Height"]))
num_feat
```

Test for Normal Distribution

```python
from scipy.stats import normaltest
```

```python
for i in num_feat:
  stats, pval=normaltest(df[i])
  if pval > 0.05:
    print(i,': Normal Distributed')
  else:
    print(i,': Not Normal Distributed')
```

Defining Y and X

```python
y = df['Absenteeism time in hours']
y
```

```python
x = df.drop(columns=["Absenteeism time in hours", "ID","Reason for absence","Seasons","Hit target","Weight","Height"])
x
```

Standardize the Inputs

```python
# prompt: standardize the x

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
x_scaled
```

Training ML model

```python
pip install scikit-learn
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(
    x,
    y,
    test_size=0.2,
    random_state=42
)
```

print(x_train.shape, x_test.shape, y_train.shape, y_test.shape)


Run Linear Regression


```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train, y_train)
```

Showing the R2 Score of the Model

```python
from sklearn.metrics import r2_score

y_pred = model.predict(x_test)
r2_score = r2_score(y_test, y_pred)
print("R2 Score:", r2_score)
```

Showing the Mean Absolute Error of the Model

```python
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(y_test, y_pred)
print("Mean Absolute Error:", mae)
```

Showing the Coefficient of the Model

```python
coef_df = pd.DataFrame({
    "feature_names":list(x_train),
    "coef": model.coef_
})
print(coef_df)
```

```python
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

Displaying the Linear Regression Equation of the Model

```python
intercept = model.intercept_
coef = model.coef_
equation = f"y = {intercept:.2f}"
for i, c in enumerate(coef):
  if c != 0:
    sign = "+" if c > 0 else "-"
    equation += f" {sign} {abs(c):.2f} * {list(x_train)[i]}"
print(equation)
```
